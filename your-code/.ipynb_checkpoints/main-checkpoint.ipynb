Complete the following set of exercises to solidify your knowledge of parsing RSS feeds and extracting information from them.

```python
import feedparser
```

### 1. Use feedparser to parse the following RSS feed URL.

```python
url = 'http://feeds.feedburner.com/oreilly/radar/atom'
```

```python
feedburner= feedparser.parse(url)
```

### 2. Obtain a list of components (keys) that are available for this feed.

```python
list_feedburner=list(feedburner.keys())
```

```python
list_feedburner
```

### 3. Obtain a list of components (keys) that are available for the *feed* component of this RSS feed.

```python
list_feed=list(feedburner.feed.keys())
```

```python
list_feed
```

### 4. Extract and print the feed title, subtitle, author, and link.

```python
print(feedburner.feed.title)
print(feedburner.feed.subtitle)
print(feedburner.feed.link)
#En el feed no existe un dato denominado 'author', razón por la que no se extrajo
```

### 5. Count the number of entries that are contained in this RSS feed.

```python
print(len(feedburner.entries))
```

### 6. Obtain a list of components (keys) available for an entry.

*Hint: Remember to index first before requesting the keys*

```python
list_entry=list(feedburner.entries[0].keys())
list_entry
```

### 7. Extract a list of entry titles.

```python
entry_titles = [feedburner.entries[i].title for i in range(len(feedburner.entries))]
entry_titles
```

### 8. Calculate the percentage of "Four short links" entry titles.

```python
number_four=0
for i in range(len(entry_titles)):
    if entry_titles[i].startswith('Four short links'):
        number_four=number_four+1
#number_four2= +1 if entry_titles[i].startswith('Four short links') for i in range(len(entry_titles))

print('The percentage of "Four short links" is:',number_four/len(entry_titles))
```

### 9. Create a Pandas data frame from the feed's entries.

```python
import pandas as pd
```

```python
df = pd.DataFrame(feedburner.entries)
df.head(3)
```

### 10. Count the number of entries per author and sort them in descending order.

```python
df.author.value_counts().sort_values(ascending=False)
```

### 11. Add a new column to the data frame that contains the length (number of characters) of each entry title. Return a data frame that contains the title, author, and title length of each entry in descending order (longest title length at the top).

```python
df = pd.DataFrame(feedburner.entries)
df['Title_lenght']=df['title'].apply(len)
df[['title','author','Title_lenght']].sort_values(by=('Title_lenght'))
```

### 12. Create a list of entry titles whose summary includes the phrase "machine learning."

```python
df = pd.DataFrame(feedburner.entries)
df["Indexes"]= df["summary"].str.find('machine learning') 
newdf = df[(df.Indexes != -1)].sort_values(by=('Indexes'), ascending=False)
list_machLea=list(newdf.title)
print(list_machLea)
```
